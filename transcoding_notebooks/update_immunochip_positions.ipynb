The files located in /data/ADI/Data_Warehouse/Genetic_Data/ImmunoChip_Data appear to have coordinates that match hg18/NCBI36.  Sure would be nice to have everything with matching coordinates, so I am going to uplift this to *at least* GRCh37

## Step 1: Retrieve hg18 to hg19 chain file:

```
curl http://hgdownload.soe.ucsc.edu/goldenPath/hg18/liftOver/hg18ToHg19.over.chain.gz -o /Volumes/guth_aci_informatics/references/hg18ToHg19.over.chain.gz
```

## Step 2: Create map file

`liftOver`, and by extension by tool [`PLINKliftOver`](https://github.com/milescsmith/PlinkLiftOver) operates on a map file, which is something generated by PLINK version 1, not version 2

```
plink --bfile CE --recode --out updating
```

## Step 3: Liftover map file

```
plinkliftover updating.map /Volumes/guth_aci_informatics/references/hg18ToHg19.over.chain --prefix hg19_CE
```

## Step 4: Update the coordinate mapping and create new set of PLINK files

```
plink2 --bfile CE --update-map updating.map 4 2 --make-pgen --sort-vars --out uplifted_CE
```
