```python
import pandas as pd

def commonRows(df1: pd.DataFrame, df2: pd.DataFrame) -> pd.Index:
  return pd.Index.intersection(df1.index, df2.index)
```

```python
import numpy as np

rng = np.random.default_rng()

mat1 = rng.random((10,10))
mat2 = rng.random((10,10))
```

```python
import string

df1 = pd.DataFrame(data=mat1, index = [_ for _ in string.ascii_letters[0:10]])
df2 = pd.DataFrame(data=mat2, index = [_ for _ in string.ascii_letters[5:15]])
```

```python
df1
```

```python
commonRows(df1, df2)
```
