```python
long_df = pd.DataFrame.from_dict(
    [
        {"pathway": y[0], "url": y[1], "genes": y[2].split("\t")}
        for y in [x.split("\t", maxsplit=2) for x in c2]
    ]
).drop(columns=["url"])
```

```python
long_df.head()
```

```python
explode_df = long_df.explode(column="genes").assign(
    genes=lambda x: x["genes"].str.strip(), count=1
)
```

```python
explode_df
```

```python
explode_df["genes"] = explode_df["genes"].str.strip()
```

```python
explode_df["count"] = 1
```

```python
explode_df
```

```python
reactome_wide = pd.pivot_table(
    explode_df, values="count", index="genes", columns="pathway", fill_value=0
)
```
