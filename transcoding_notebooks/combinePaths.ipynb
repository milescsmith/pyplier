```python
import pandas as pd
from pathlib import Path
```

```python
dataPath = Path.cwd().parent / "tests" / "data" / "combine_paths"

allPaths = pd.read_csv(dataPath.joinpath("allPaths.csv"), index_col=0)
bloodCellMarkers = pd.read_csv(
    dataPath.joinpath("bloodCellMarkersIRISDMAP.csv"), index_col=0
)
canonicalPathways = pd.read_csv(dataPath.joinpath("canonicalPathways.csv"), index_col=0)
```
