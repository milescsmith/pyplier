```python
from copy import deepcopy
from typing import Optional

import numpy as np
import pandas as pd
from glmnet import ElasticNet
```

```python
glm_alpha = 0.9
maxPath = 10
target_frac = 0.7
L3 = None
pathwaySelection = "complete"

Ur = Chat @ Z  # get U by OLS

Ur = Ur.rank(axis="index", ascending=False)  # rank
Urm = Ur.min(axis=1)
```

```python
lambdas = np.exp(np.arange(start=-4, stop=-12.125, step=-0.125))
results = dict()
lMat = np.full((len(lambdas), Z.shape[1]), np.nan)
```

```python
pathwaySelection = "fast"
# U = np.zeros(shape = (priorMat.shape[1], Z.shape[1]))if L3 is None:
lambdas = np.exp(np.arange(start=-4, stop=-12.125, step=-0.125))
results = dict()
lMat = np.zeros((len(lambdas), Z.shape[1]))

for x in range(Z.shape[1]):
    if pathwaySelection == "fast":
        iip = np.where([Ur.iloc[:, x] <= maxPath])[1]
    else:
        iip = np.where([Urm <= maxPath])[1]

    gres = ElasticNet(
        lambda_path=lambdas,
        lower_limits=0,
        standardize=False,
        fit_intercept=True,
        alpha=glm_alpha,
        max_features=150,
    )

    gres.fit(
        y=Z.iloc[:, x].astype(np.float64).values,
        X=priorMat.iloc[:, iip].astype(np.float64).values,
        relative_penalties=penalty_factor[iip],
    )
    gres.iip = iip
    lMat[:, x] = np.sum(np.where(gres.coef_path_ > 0, 1, 0), axis=0)
    results[x] = deepcopy(gres)

fracs = np.mean(np.where(lMat > 0, 1, 0), axis=1)
iibest = np.where((target_frac - fracs) == min(abs(target_frac - fracs)))[0][0]
```
